```python
corpus = pd.read_table('pulmonar.txt', sep = " ")
corpus
```

```python
#dividimos el texto en tokens, eliminando cualquier signo de puntuación que se encuentre junto a una palabra
tokenizer = RegexpTokenizer(r'\w+')
corpus['abstract'] = corpus['abstract'].apply(lambda x: tokenizer.tokenize(x.lower()))
corpus['abstract'].head(20)
```

```python
#eliminamos las stopwords
def remove_stopwords(text):
    words = [w for w in text if w not in stopwords.words('english')]
    return words
corpus['abstract'] = corpus['abstract'].apply(lambda x: remove_stopwords(x))
corpus['abstract'].head(20)
```

```python
#eliminamos tokens de dimension 1
def una_dimension(text):
    words = [w for w in text if len(w) != 1]
    return words
corpus['abstract'] = corpus['abstract'].apply(lambda x: una_dimension(x))
corpus['abstract'].head(20)
```

```python
#eliminamos palabras comunes
def palabras_comunes(text):
    palabras = ['background', 
                'methods', 
                'method',
                'material', 
                'materials', 
                'results', 
                'result', 
                'discussion', 
                'ii', 
                'iii', 
                'conclusion', 
                'conclusions', 
                'keywords']
    words = [w for w in text if w not in palabras]
    return words

corpus['abstract'] = corpus['abstract'].apply(lambda x: palabras_comunes(x))
corpus['abstract'].head(20)
```

```python
#eliminamos numeros que no esten asociados a ningun token
def eliminar_numeros_sueltos(text):
    words = ' '.join(w for w in text if not w.isdigit())
    return words
corpus['abstract'] = corpus['abstract'].apply(lambda x: eliminar_numeros_sueltos(x))
corpus['abstract'].head(20)

#lo volveremos a dividir si queremos hacer el paso de Stemm 

corpus['abstract'] = corpus['abstract'].apply(lambda x: tokenizer.tokenize(x.lower()))
corpus['abstract'].head(50)
```

```python
#seleccionamos la raiz de las palabras
lemmatizer = WordNetLemmatizer()

def word_lemmatizer(text):
    lem_text = [lemmatizer.lemmatize(i) for i in text]
    return lem_text

corpus['abstract'].apply(lambda x: word_lemmatizer(x))
```

```python
#quitar los tokens
stemmer = PorterStemmer()

def word_stemmer(text):
    stem_text = " ".join([stemmer.stem(i) for i in text])
    return stem_text

corpus['abstract'] = corpus['abstract'].apply(lambda x: word_stemmer(x))
corpus['abstract']
```

```python
corpus.to_csv('pulmonarprocesado.csv')
```
